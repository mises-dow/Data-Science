```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# Except for the "uniform" distribution (which is meant to show constant probability), 
# I would say the descriptive statistics across all these plots did a good job of 
# indicating where the majority of data points are distributed in a given data set. 

```

```python
# exponential distribution

exponential = np.random.exponential(5, 10000)
plt.hist(exponential)
plt.axvline(exponential.mean(), color='purple', linestyle='solid', linewidth=2)
plt.axvline(exponential.std()+exponential.mean(), color='purple', linestyle='dashed', linewidth=2)
plt.axvline(exponential.mean()-exponential.std(), color='purple', linestyle='dashed', linewidth=2)
plt.show()
```

```python
# poisson distribution
poisson = np.random.poisson(15, 100)
plt.hist(poisson)
plt.axvline(poisson.mean(), color='purple', linestyle='solid', linewidth=2)
plt.axvline(poisson.std()+poisson.mean(), color='purple', linestyle='dashed', linewidth=2)
plt.axvline(poisson.mean()-poisson.std(), color='purple', linestyle='dashed', linewidth=2)
plt.show()
```

```python
# binomial distribution 
binomial = np.random.binomial(1000, .5, 100)
plt.hist(binomial)
plt.axvline(binomial.mean(), color='purple', linestyle='solid', linewidth=2)
plt.axvline(binomial.std()+binomial.mean(), color='purple', linestyle='dashed', linewidth=2)
plt.axvline(binomial.mean()-binomial.std(), color='purple', linestyle='dashed', linewidth=2)
plt.show()
```

```python
# negative binomial distribution
neg_bin = np.random.negative_binomial(1000, .5, 100)
plt.hist(neg_bin)
plt.axvline(neg_bin.mean(), color='purple', linestyle='solid', linewidth=2)
plt.axvline(neg_bin.std()+neg_bin.mean(), color='purple', linestyle='dashed', linewidth=2)
plt.axvline(neg_bin.mean()-neg_bin.std(), color='purple', linestyle='dashed', linewidth=2)
plt.show()
```

```python
# uniform distribution
uniform = np.random.uniform(1, 50, 100)
plt.hist(uniform)
plt.axvline(uniform.mean(), color='purple', linestyle='solid', linewidth=2)
plt.axvline(uniform.std()+uniform.mean(), color='purple', linestyle='dashed', linewidth=2)
plt.axvline(uniform.mean()-uniform.std(), color='purple', linestyle='dashed', linewidth=2)
plt.show()
```

```python
# Gaussian distribution
normal = np.random.normal(0, 1, 100)
plt.hist(normal)
plt.axvline(normal.mean(), color='purple', linestyle='solid', linewidth=2)
plt.axvline(normal.std()+normal.mean(), color='purple', linestyle='dashed', linewidth=2)
plt.axvline(normal.mean()-normal.std(), color='purple', linestyle='dashed', linewidth=2)
plt.show()
```

```python
# 2 normal distributions added together
normal1 = np.random.normal(5, .5, 100)
normal2 = np.random.normal(10, 1, 100)
normal3 = np.random.normal(15, 1.5, 100)


plt.hist(normal3)
plt.axvline(normal3.mean(), color='purple', linestyle='solid', linewidth=2)
plt.axvline(normal3.std()+normal3.mean(), color='purple', linestyle='dashed', linewidth=2)
plt.axvline(normal3.mean()-normal3.std(), color='purple', linestyle='dashed', linewidth=2)
plt.show()
```
