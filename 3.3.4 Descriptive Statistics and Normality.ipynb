```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# exponential distribution

exponential = np.random.exponential(5, 100)
plt.hist(exponential)
plt.axvline(exponential.mean(), color='purple', linestyle='solid', linewidth=2)
plt.axvline(exponential.std()+exponential.mean(), color='purple', linestyle='dashed', linewidth=2)
plt.axvline(exponential.std()-exponential.mean(), color='purple', linestyle='dashed', linewidth=2)
plt.show()
```

```python
# beta distribution 
beta = np.random.beta(.05, .5, 100)
plt.hist(beta)
plt.axvline(beta.mean(), color='purple', linestyle='solid', linewidth=2)
plt.axvline(beta.std()+beta.mean(), color='purple', linestyle='dashed', linewidth=2)
plt.axvline(beta.std()-beta.mean(), color='purple', linestyle='dashed', linewidth=2)
plt.show()
```

```python
# poisson distribution
poisson = np.random.poisson(15, 100)
plt.hist(poisson)
plt.axvline(poisson.mean(), color='purple', linestyle='solid', linewidth=2)
plt.axvline(poisson.std()+poisson.mean(), color='purple', linestyle='dashed', linewidth=2)
plt.axvline(poisson.std()-poisson.mean(), color='purple', linestyle='dashed', linewidth=2)
plt.show()
```

```python
# dirichlet distribution
dirichlet = np.random.dirichlet((2, 4, 8), 100)
plt.hist(dirichlet)
plt.axvline(dirichlet.mean(), color='purple', linestyle='solid', linewidth=2)
plt.axvline(dirichlet.std()+dirichlet.mean(), color='purple', linestyle='dashed', linewidth=2)
plt.axvline(dirichlet.std()-dirichlet.mean(), color='purple', linestyle='dashed', linewidth=2)
plt.show()
```

```python
# uniform distribution
uniform = np.random.uniform(1, 1000, 100)
plt.hist(uniform)
plt.axvline(uniform.mean(), color='purple', linestyle='solid', linewidth=2)
plt.axvline(uniform.std()+uniform.mean(), color='purple', linestyle='dashed', linewidth=2)
plt.axvline(uniform.std()-uniform.mean(), color='purple', linestyle='dashed', linewidth=2)
plt.show()
```

```python
# geometric distribution
geometric = np.random.geometric(.25, 100)
plt.hist(geometric)
plt.axvline(geometric.mean(), color='purple', linestyle='solid', linewidth=2)
plt.axvline(geometric.std()+geometric.mean(), color='purple', linestyle='dashed', linewidth=2)
plt.axvline(geometric.std()-geometric.mean(), color='purple', linestyle='dashed', linewidth=2)
plt.show()
```
