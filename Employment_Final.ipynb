```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('employment_2016.csv').dropna()


df['Occupations'] = df['occ title'].apply(lambda title: title[:21])

new_york = df[df['area_title']== 'New York']
top_10_ny = new_york.nlargest(20, 'mean_annual_wage') 
new_jersey = df[df['area_title']== 'New Jersey']
top_10_nj = new_jersey.nlargest(20, 'mean_annual_wage')

```

```python
df.head(10)
```

```python
text1 = 'Here we can see NY Wages'
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8,4), sharey=True)
top_10_ny.plot.bar(x='Occupations', y='mean_annual_wage', ax=ax1)
top_10_nj.plot.bar(x='Occupations', y='mean_annual_wage', ax=ax2)
ax1.set_title('Top 20 Jobs by Wage, NY')
ax1.set_ylabel('Wages')
ax2.set_title('Top 20 Jobs by Wage, NJ') 
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8,4))
new_york.plot.scatter(x='mean_hourly_wage', y='mean_annual_wage', ax= ax1)
ax1.set_ylabel('All Wages in NY')
ax1.set_title('Mean Wages in NY')

new_jersey.plot.scatter(x='mean_hourly_wage', y='mean_annual_wage', ax= ax2)
ax2.set_ylabel('All Wages in NJ')
ax2.set_title('Mean Wages in NJ')

plt.tight_layout()
```
